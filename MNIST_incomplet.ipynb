```python
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Import the MNIST dataset directly from Keras.
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data(path="mnist.npz")

# Transform the data shape to (28, 28, 1)
# Then normalize the data to values between 0 and 1
# because neural network functions best within this range
X_train, X_test = X_train[..., np.newaxis] / 255.0, X_test[..., np.newaxis] / 255.0
print(X_train.shape, X_test.shape)
```

```python
# Construisez et compilez votre modèle ici
model = tf.keras.models.Sequential([
#         tf.keras.layers.Conv2D(28, (3, 3),
#                       activation='relu', padding='same', input_shape=(28, 28, 1))(X_train_T),
        tf.keras.layers.Input(shape=(28, 28, 1)),
        tf.keras.layers.Conv2D(8, 3, activation='relu', padding="same"),
        tf.keras.layers.Flatten(input_shape=(28, 28, 1)),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(10, activation='linear'),
])
```

```python
# "Compilation" du modèle: On indique quelle fonction de perte et quel optimisateur utiliser.
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), # "Mean squared error", perte quadratique
    metrics=["accuracy"]
)
```

```python
# Affichage de la structure du modèle
model.summary()
```

```python
# Entraînez votre modèle ici
model.fit(X_train, y_train, epochs=10)
```

```python
# Une fois satisfaits de votre performance sur le validation set, évaluez une seule fois sur le test set
model.evaluate(X_test, y_test)
```

```python
# Sauvegardez bien votre modèle
model.save("models/mnist_model.h5")
```
